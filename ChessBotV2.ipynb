```python
class Chess:
    def __init__(self):
      self.board = [
              ['Rb', 'Nb', 'Bb', 'Qb', 'Kb', 'Bb', 'Nb', 'Rb'],
              ['Pb', 'Pb', 'Pb', 'Pb', 'Pb', 'Pb', 'Pb', 'Pb'],
              ['  ', '  ', '  ', '  ', '  ', '  ', '  ', '  '],
              ['  ', '  ', '  ', '  ', '  ', '  ', '  ', '  '],
              ['  ', '  ', '  ', '  ', '  ', '  ', '  ', '  '],
              ['  ', '  ', '  ', '  ', '  ', '  ', '  ', '  '],
              ['Pw', 'Pw', 'Pw', 'Pw', 'Pw', 'Pw', 'Pw', 'Pw'],
              ['Rw', 'Nw', 'Bw', 'Qw', 'Kw', 'Bw', 'Nw', 'Rw']]
      self.player = 'white' 
      self.checkmate = False
      #self.sample = ["a2a4","b7b5","a1a3","b5a4"]#Pawn Test
      #self.sample = ["a2a4","a7a5","a1a3","a8a6","a3d3","a3d3","a6d6","c2d3","c2c4","d6d3","e2d3"]#Rook Test
      #self.sample  = ["b2b4","d7d5","c1a3","c8g4","b4b5","g4e2","a3e7"]#Bishop Test
      #self.sample = ["b1c3","d7d5","c3d5","c8g4","b4b5","d5f6","e7e6","f6g4"]#Knight Test
      #self.sample = ["d2d4","d7d5","d1d3","d8d6","d3c3","d6c6","c3c6","b7b6","c6d5"]#Queen Test
      #self.sample = ["b1c3","d7d5","c3d5","c8g4","d5f6","d8d6","e8d7"]#Knight Check King Test 
      #self.sample = ["c2c4","d7d5","c4d5","e7e5","d5d6","e5e4","d6d7","e4e3","d8d7","f2f3","e4f3","e2e4","f3f2","b2b3"]#Pawn Check King Test 
      #self.sample  = ["b2b4","d7d5","c1a3","c8g4","b4b5","g4e2","a3e7","d8d6","e7f8","d6e6","e1e2","a2a3","e2f3","a3a4","f1e2","e6e2","a3a4","d1e2","f3e2"]#Before and after move Check King Test 
      #self.sample = ["a2a4","b7b5","a4a6","b2b4","c7c5","c2c4","d7d5","d2d4","e7e5","e2e4","e7e5"]
      self.sample = []
      self.count = 0

    def print_board(self, board):
        count = 8
        print('     a   b   c   d   e   f   g   h\n--------------------------------------')
        for row in board:
            print(count," ",end="|")
            print(' |'.join(row),"|",count)
            print('--------------------------------------')
            count -=1
        print('     a   b   c   d   e   f   g   h')
        
    def get_move(self, board):
        valid_input = False
        while not valid_input:
            if self.player == 'white':
                move = input(f"{self.player}'s turn. Enter your move (e.g. e2e4): ")
            else:
                move = self.sample[self.count]
            print(move)
            
            if len(move) == 4 and move[0] in 'abcdefgh' and move[1] in '12345678' and move[2] in 'abcdefgh' and move[3] in '12345678':
                valid_input = True
                move = 8 - int(move[1]),ord(move[0]) - ord('a'), 8 - int(move[3]), ord(move[2]) - ord('a') # start row, start col, end row , end col
                print(move)
                if self.count < len(self.sample):
                    self.count+=1
                #self.sample.append(move)
                
            else:
                print("Invalid input. Try like (e2e4) => (start row, start col, end row , end col)")
            
        return move


    def is_valid_move(self, move, justChecking, board, demo = False):
      sr, sc, er, ec = move[0],move[1],move[2],move[3]
      piece = board[sr][sc]
      flag = 0
      if not ( 0 <= er < 8 and 0 <= ec < 8):return False
      if(piece == '  '):
        print("There is no piece for you")
        return False
      #Check if current player is giving the input
      if ((piece.find('b') == -1 and self.player == 'black') or \
          (piece.find('w') == -1 and self.player == 'white')) and justChecking == False:
        print("Give chance to other player. It's not your turn")
        #print(move)
        return False
      #Check if same player taking its own piece
      if((board[er][ec].find('b') !=-1 and self.player == 'black') or (board[er][ec].find('w') !=-1 and self.player == 'white')) and justChecking == False:
        #print(board[er][ec].find('b'), self.player)
        print("Seriously ? You can't take your own piece")
        return False

      # White pawn moves
      if piece == 'Pw':
        if board[er][ec] == '  ':
          if (sr == er+1 and sc == ec ) or (sr == er+2 and sc == ec and sr > 5):
            flag = 1
      # When White pawn can capture
        elif ((sr == er+1 and sc == ec+1) or (sr == er+1 and sc == ec-1)) and board[er][ec].find('w') == -1:
            
            flag = 1
      # Black pawn moves
      elif piece == 'Pb' :
        if board[er][ec] == '  ':
          if (sr == er-1 and sc == ec ) or (sr == er-2 and sc == ec and sr < 2):
              
            flag = 1
      # When Black pawn can capture
        elif ((sr == er-1 and sc == ec+1) or (sr == er-1 and sc == ec-1)) and board[er][ec].find('b') == -1:
            flag = 1
      # Rook moves
      if piece == 'Rw' or  piece == 'Rb':
        if(self.clear(move,piece, board)):
         if (piece == "Rw" and board[er][ec].find("w") == -1) or (piece == "Rb" and board[er][ec].find("b") == -1):
          flag = 1
      if piece == 'Bw' or piece == 'Bb':
        if(self.clear(move,piece, board)):
          flag = 1
      if piece == 'Nw' or piece == 'Nb':
        if abs(er - sr) == 2 and abs(ec - sc) == 1 or abs(er - sr) == 1 and abs(ec - sc) == 2:
            if(piece == 'Nb' and board[er][ec].find('b') == -1) or (piece == 'Nw' and board[er][ec].find('w') == -1):
                flag = 1
      if piece == 'Qw' or piece == 'Qb':
        if(self.clear(move,piece, board)):
          flag = 1
      #King Move
      if piece == 'Kw' or piece == 'Kb':
        if abs(er - sr) <= 1 and abs(ec - sc) <= 1:
          move,taken_piece,who_took_piece = self.move(move,board)
          if not self.king_safety(self.player,board):
            self.moveBack(move,taken_piece,who_took_piece,board)
            #self.print_board()
            print("King Can't Move there")
            return False
          else:
            self.moveBack(move,taken_piece,who_took_piece,board)
            #self.print_board()
            flag = 1
      if flag == 1:
        #if demo == True and self.point(board,move) != 0:print(self.point(board,move))
        return True
      else:
        return False


    def clear(self,move,piece,board):
      sr, sc, er, ec = move[0],move[1],move[2],move[3]
      #print("Hi2",sr,sc,er, ec, piece)
      
      # Rook Check
      if(piece == 'Rb' or piece == 'Rw' or piece == 'Qb' or piece == 'Qw'):
        if sr == er:
          #print("Hi",3)
          for i in range(min(sc,ec),max(sc,ec)):
            if(board[sr][i] != '  ' and i != sc and i != ec):
             #print(1)
              return False
          return True
        elif sc == ec:
          #print("Hi",4)
          for i in range(min(sr,er),max(sr,er)):
            #print(board[i][sc])
            if(board[i][sc] != '  ' and i != sr and i != er):
              return False
          return True
      if(piece == 'Bb' or piece == 'Bw' or piece == 'Qb' or piece == 'Qw'):
        if abs(sr-er) == abs(sc-ec) and ((sr>er and sc > ec) or (sr < er and sc < ec)):
          j = min(sc,ec)
          for i in range(min(sr,er),max(sr,er)):
              if( board[i][j] != '  ' and i != sr and i != sc ):
                  return False
              j+=1
          return True
        #print(123123)
        if sr+sc == er+ec:
          #print("Hi",6)
          for i in range(min(sr,er),max(sr,er)):
            for j in range(min(sc,ec),max(sc,ec)):
              #print(board[i][j],i,j,i+j,sr+er)
              if i+j == sr+sc:
                if( board[i][j] != '  ' and i != sr and i != sc ):
                  #print(11)
                  return False
          return True
      return False


    def move(self,move, board):
      sr, sc, er, ec = move[0],move[1],move[2],move[3]
      taken_piece = board[er][ec]
      who_took_piece = board[sr][sc]
      board[er][ec] = board[sr][sc]
      board[sr][sc] = '  '
      return move,taken_piece,who_took_piece


    #If king is checked by X, it help to move other piece of king to capture X  
    def moveBack(self,move,taken_piece,who_took_piece,board):
      sr, sc, er, ec = move[0],move[1],move[2],move[3]
      board[sr][sc] = who_took_piece
      board[er][ec] = taken_piece

    def king_find(self,board):
      if( self.player == 'black'):
        for i in range(8):
          for j in range(8):
            #print(i,j,self.board[i][j])
            if(board[i][j] == 'Kb'):
              #print(i,j)
              return i,j
      else:
        for i in range(8):
          for j in range(8):
            #print(i,j,self.board[i][j])
            if(board[i][j] == 'Kw'):
              #print(i,j)
              return i,j


    def king_safety(self,attack,board):
      self.print_board(board)
      #print(attack)
      
      kr,kc = self.king_find(board)
      # Check for attacks by pawns
      if attack == 'white':
          pa = [(kr+1, kc-1), (kr+1, kc+1)]
          for row, col in pa:
            #print(kr+1,kc-1,self.board[row][col])
            if 0 <= row < 8 and 0 <= col < 8 and board[row][col].find('Pw') !=-1:
                return False
      else:
          pa = [(kr-1, kc-1), (kr-1, kc+1)]
          for row, col in pa:
            if 0 <= row < 8 and 0 <= col < 8 and board[row][col].find('Pb') !=-1:
              return False

      # Check for attacks by knights
      nm = [(2,1), (1,2), (-1,2), (-2,1), (-2,-1), (-1,-2), (1,-2), (2,-1)] # Knight Moves
      if attack == 'white':
        for row, col in [(kr+r, kc+c) for r, c in nm]:
            if 0 <= row < 8 and 0 <= col < 8 and board[row][col].find('Nw') != -1 :
              return False
      if attack == 'white':
        for row, col in [(kr+r, kc+c) for r, c in nm]:
            if 0 <= row < 8 and 0 <= col < 8 and board[row][col].find('Nb') != -1 :
              return False
      #Check for attacks by Bishop

      if attack == 'white':
        for i in range(8):
          for j in range(8):
            if (board[i][j] == 'Bw' or board[i][j] == 'Rw' or board[i][j] == 'Qw') and\
                self.is_valid_move([i,j,kr,kc],True,board):
              #print(i,j,kr,kc,board[i][j],board[kr][kc],self.is_valid_move([i,j,kr,kc],True,board))
              
              #print(123123123213)
              return False
      if attack == 'black':
        for i in range(8):
          for j in range(8):
            
            if (board[i][j] == 'Bb' or board[i][j] == 'Rb' or board[i][j] == 'Qb') and\
                self.is_valid_move([i,j,kr,kc],True,board):
              #print(9999999999999)
              return False
      return True
  
    
  
    #Pawn: 1 point (or pawn)
    #Knight: 3 points
    #Bishop: 3 points
    #Rook: 5 points
    #Queen: 9 points
    
    
    def point(self, board, move):
      white = 0
      black = 0
      move,taken_piece,who_took_piece = self.move(move,board)
      for i in range(8):
         for j in range(8):
           if(board[i][j].find('w')) !=-1:
             if(board[i][j].find('P'))!=-1:
               white = white + 1
             elif(board[i][j].find('N')) !=-1:
               white = white + 3
             elif(board[i][j].find('B'))!=-1:
               white = white + 3
             elif(board[i][j].find('R'))!=-1:
               white = white + 5
             elif(board[i][j].find('Q'))!=-1:
               white = white + 9
         
           elif(board[i][j].find('b'))!=-1:
             if(board[i][j].find('P'))!=-1:
               black = black + 1
             elif(board[i][j].find('N'))!=-1:
               black = black + 3
             elif(board[i][j].find('B'))!=-1:
               black = black + 3
             elif(board[i][j].find('R'))!=-1:
               black = black + 5
             elif(board[i][j].find('Q'))!=-1:
               black = black + 9
      self.moveBack(move,taken_piece,who_took_piece,board)
      return white-black
  
    
  
    def move_to_word(self,i):
        return chr(ord('a')+i[1])+str(8-i[0])+chr(ord('a')+i[3])+str(8-i[2])
    
    def all_moves(self, board):
        all_moves = []
        move_serial = []
        point_per_move = []
        for row in range(8):
            for col in range(8):
                piece = board[row][col]
                if piece == "Pb":
                    if(self.is_valid_move([row,col,row+1,col],True,board)):
                      all_moves.append([row,col,row+1,col])
                      move_serial.append("b")
                    if(self.is_valid_move([row,col,row+2,col],True,board)):
                      all_moves.append([row,col,row+2,col])
                      move_serial.append("b")
                    if(self.is_valid_move([row,col,row+1,col+1],True,board)):
                      all_moves.append([row,col,row+1,col+1])
                      move_serial.append("b")
                    if(self.is_valid_move([row,col,row+1,col-1],True,board)):
                      all_moves.append([row,col,row+1,col-1])
                      move_serial.append("b")
                if piece == "Pw":
                    if(self.is_valid_move([row,col,row-1,col],True,board)):
                      all_moves.append([row,col,row-1,col])
                      move_serial.append("w")
                    if(self.is_valid_move([row,col,row-2,col],True,board)):
                      all_moves.append([row,col,row-2,col])
                      move_serial.append("w")
                    if(self.is_valid_move([row,col,row-1,col+1],True,board)):
                      all_moves.append([row,col,row-1,col+1])
                      move_serial.append("w")
                    if(self.is_valid_move([row,col,row-1,col-1],True,board)):
                      all_moves.append([row,col,row-1,col-1])
                      move_serial.append("w")
                if piece == "Nw" or piece == "Nb":
                  nm = [(2,1), (1,2), (-1,2), (-2,1), (-2,-1), (-1,-2), (1,-2), (2,-1)]
                  for Erow, Ecol in nm:
                    if(self.is_valid_move([row,col,row+Erow,col+Ecol],True,board)):
                      if board[row][col] == "Nw":
                        all_moves.append([row,col,row+Erow,col+Ecol])
                        move_serial.append("w")
                      else:
                        all_moves.append([row,col,row+Erow,col+Ecol])
                        move_serial.append("b")
                if piece == "Rb" or piece == "Rw":
                    for i in range(8):
                        if(self.is_valid_move([row,col,i,col],True,board)):
                            all_moves.append([row,col,i,col])
                            move_serial.append("b") if piece == "Rb" else move_serial.append("w")
                        if(self.is_valid_move([row,col,row,i],True,board)):
                            all_moves.append([row,col,row,i])
                            move_serial.append("b") if piece == "Rb" else move_serial.append("w")
        print("Valid move count:",len(all_moves))

        for j in range(len(all_moves)):
          i = all_moves[j]
          
          
          if(self.player == 'white' and move_serial[j] == 'w') or \
           (self.player == 'black' and move_serial[j] == 'b'):
            #print(i)
            self.is_valid_move(i,True,board,True)
            #print(chr(ord('a')+i[1])+str(8-i[0])+chr(ord('a')+i[3])+str(8-i[2]),move_serial[j])
            move_in_word = chr(ord('a')+i[1])+str(8-i[0])+chr(ord('a')+i[3])+str(8-i[2])
            point_per_move.append([self.point(board, i),move_in_word,move_serial[j]])
        min = 100
        max = -100
        min_move = ''
        max_move = ''
        for i in point_per_move:
            print(i)
            if i[0] < min :
                min = i[0]
                min_move = i[1]
            print(i)
            if i[0] > max :
                max = i[0]
                max_move = i[1]
        print("Best Move for", self.player)
        print(min_move) if self.player == 'black' else print(max_move)
        print(self.sample)
        self.sample.append(min_move)
        return all_moves


    def play(self):
      while not self.checkmate:
        self.all_moves(self.board)
        self.print_board(self.board)
        move = self.get_move(self.board)
        isKingMove = True if self.board[move[0]][move[1]].find("Kw") != -1 and\
         self.board[move[0]][move[1]].find("Kb") != -1 else False
        if not self.is_valid_move(move,isKingMove, self.board):
          #print(123)
          print("Invalid move")
          continue
        move,taken_piece,who_took_piece = self.move(move,self.board)
        if not (self.king_safety("black" if self.player == "white" else "white",self.board)) :
          self.moveBack(move,taken_piece,who_took_piece)
          print("King will be unsafe")
          continue
        self.player = 'black' if self.player == 'white' else 'white'
        

game = Chess()
game.play()
```
