```python
#@title Default title text
import math
class Chess:
    def __init__(self):
      self.board = [
              ['Rb', 'Nb', 'Bb', 'Qb', 'Kb', 'Bb', 'Nb', 'Rb'],
              ['Pb', 'Pb', 'Pb', 'Pb', 'Pb', 'Pb', 'Pb', 'Pb'],
              ['  ', '  ', '  ', '  ', '  ', '  ', '  ', '  '],
              ['  ', '  ', '  ', '  ', '  ', '  ', '  ', '  '],
              ['  ', '  ', '  ', '  ', '  ', '  ', '  ', '  '],
              ['  ', '  ', '  ', '  ', '  ', '  ', '  ', '  '],
              ['Pw', 'Pw', 'Pw', 'Pw', 'Pw', 'Pw', 'Pw', 'Pw'],
              ['Rw', 'Nw', 'Bw', 'Qw', 'Kw', 'Bw', 'Nw', 'Rw']]
      self.player = 'white' 
      self.checkmate = False
      #self.sample = ["a2a4","a7a5","a1a3","a8a6","a3d3","a3d3","a6d6","c2d3","c2c4","d6d3","e2d3"]#Rook Test
      #self.sample  = ["b2b4","d7d5","c1a3","c8g4","b4b5","g4e2","a3e7"]#Bishop Test
      #self.sample = ["b1c3","d7d5","c3d5","c8g4","b4b5","d5f6","e7e6","f6g4"]#Knight Test
      #self.sample = ["d2d4","d7d5","d1d3","d8d6","d3c3","d6c6","c3c6","b7b6","c6d5"]#Queen Test
      self.sample = ["b1c3","d7d5","c3d5","c8g4","b4b5","d5f6","e7e6",""]#Knight Test
      self.count = 0

    def print_board(self):
        count = 8
        print('     a   b   c   d   e   f   g   h\n--------------------------------------')
        for row in self.board:
            print(count," ",end="|")
            print(' |'.join(row),"|",count)
            print('--------------------------------------')
            count -=1
        print('     a   b   c   d   e   f   g   h')
        
    def get_move(self):
        valid_input = False
        while not valid_input:
            #move = input(f"{self.player}'s turn. Enter your move (e.g. e2e4): ")
            move = self.sample[self.count]
            print(move)
            if self.count < len(self.sample):
                self.count+=1
            if len(move) == 4 and move[0] in 'abcdefgh' and move[1] in '12345678' and move[2] in 'abcdefgh' and move[3] in '12345678':
                valid_input = True
                move = 8 - int(move[1]),ord(move[0]) - ord('a'), 8 - int(move[3]), ord(move[2]) - ord('a') # start row, start col, end row , end col
                print(move)
            else:
                print("Invalid input. Try like (e2e4) => (start row, start col, end row , end col)")
        return move


    def is_valid_move(self, move):
      sr, sc, er, ec = move[0],move[1],move[2],move[3]
      piece = self.board[sr][sc]
      #Check if current player is giving the input
      if (piece.find('b') == -1 and self.player == 'black') or (piece.find('w') == -1 and self.player == 'white'):
        print("Give chance to other player. It's not your turn")
        return False
      #Check if same player taking its own piece
      if(self.board[er][ec].find('b') !=-1 and self.player == 'black') or (self.board[er][ec].find('w') !=-1 and self.player == 'white'):
        print(self.board[er][ec].find('b'), self.player)
        print("Seriously ? You can't take your own piece")
        return False
      # White pawn moves
      if piece == 'Pw':
        if self.board[er][ec] == '  ':
          if (sr == er+1 and sc == ec ) or (sr == er+2 and sc == ec):
            return True
      # When White pawn can capture
        elif (sr == er+1 and sc == ec+1) or (sr == er+1 and sc == ec-1):
            return True
      # Black pawn moves
      elif piece == 'Pb' :
        if self.board[er][ec] == '  ':
          if (sr == er-1 and sc == ec ) or (sr == er-2 and sc == ec):
            return True
      # When Black pawn can capture
        elif (sr == er-1 and sc == ec+1) or (sr == er-1 and sc == ec-1):
            return True
      # Rook moves
      if piece == 'Rw' or  piece == 'Rb':
        if(self.clear(move,piece)):
          return True
      if piece == 'Bw' or piece == 'Bb':
        if(self.clear(move,piece)):
          return True
      if piece == 'Nw' or piece == 'Nb':
        if abs(er - sr) == 2 and abs(ec - sc) == 1 or abs(er - sr) == 1 and abs(ec - sc) == 2:
          return True
      if piece == 'Qw' or piece == 'Qb':
        if(self.clear(move,piece)):
          return True
      return False

    def clear(self,move,piece):
      sr, sc, er, ec = move[0],move[1],move[2],move[3]
      # Rook Check
      if(piece == 'Rb' or piece == 'Rw' or piece == 'Qb' or piece == 'Qw'):
        if sr == er:
          #print(3)
          for i in range(min(sc,ec),max(sc,ec)):
            if(self.board[sr][i] != '  ' and i != sc and i != ec):
             #print(1)
              return False
          return True
        elif sc == ec:
          #print(4)
          for i in range(min(sr,er),max(sr,er)):
            #print(self.board[i][sc])
            if(self.board[i][sc] != '  ' and i != sr and i != er):
              return False
          return True
      if(piece == 'Bb' or piece == 'Bw' or piece == 'Qb' or piece == 'Qw'):
        if abs(sr-sc) == abs(er-ec):
          for i in range(sr,er):
            for j in range(sc,ec):
              #print(self.board[i][j],i,j,abs(i-j),abs(sr-sc))
              if abs(i-j) == abs(sr-sc):
                if( self.board[i][j] != '  ' and i != sr and i != sc ):
                  return False
          return True
        #print(123123)
        if sr+sc == er+ec:
          for i in range(min(sr,er),max(sr,er)):
            for j in range(min(sc,ec),max(sc,ec)):
              #print(self.board[i][j],i,j,i+j,sr+er)
              if i+j == sr+sc:
                if( self.board[i][j] != '  ' and i != sr and i != sc ):
                  #print(11)
                  return False
          return True
      return False


    def move(self,move):
      sr, sc, er, ec = move[0],move[1],move[2],move[3]
      self.board[er][ec] = self.board[sr][sc]
      self.board[sr][sc] = '  '


    def king_find(self):
      if( self.player == 'black'):
        for i in range(8):
          for j in range(8):
            #print(i,j,self.board[i][j])
            if(self.board[i][j] == 'Kb'):
              return i,j
      else:
        for i in range(8):
          for j in range(8):
            #print(i,j,self.board[i][j])
            if(self.board[i][j] == 'Kw'):
              return i,j


    def king_safety(self,attack):
      kr,kc = self.king_find()
      # Check for attacks by pawns
      if attack == 'white':
          pa = [(kr-1, kc-1), (kr-1, kc+1)]
      else:
          pa = [(kr+1, kc-1), (kr+1, kc+1)]
      for row, col in pa:
          if 0 <= row <= 7 and 0 <= col <= 7 and self.board[row][col].find('P') and self.board[row][col].find('w') != (attack == 'white'):
              return False

      # Check for attacks by knights
      nm = [(2,1), (1,2), (-1,2), (-2,1), (-2,-1), (-1,-2), (1,-2), (2,-1)] # Knight Moves
      for row, col in [(kr+r, kc+c) for r, c in nm]:
          if 0 <= row <= 7 and 0 <= col <= 7 and self.board[row][col].find('N') and self.board[row][col].find('w') != (attack == 'white'):
              return False
      return True
      

    def play(self):
      while not self.checkmate:
        self.print_board()
        move = self.get_move()
        if not self.is_valid_move(move):
          #print(123)
          print("Invalid move")
          continue
        attack = 'black' if self.player == 'white' else 'white'
        if self.king_safety(attack):
          print("If king is unsafe what is the reason to continue ?")
          continue
        self.move(move)
        
        self.player = 'black' if self.player == 'white' else 'white'

game = Chess()
game.play()






```
